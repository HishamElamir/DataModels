```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import LabelEncoder

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
%matplotlib inline
sns.set()
```

```python
data = load_iris()
df = pd.DataFrame(data.data)
df.columns = data.feature_names
df['Iris.type'] = data.target
```

```python
cols_names = data.feature_names
```

```python
le = LabelEncoder()
le.fit(df['Iris.type'])
le.classes_ = data.target_names
df['Iris.cat_type'] = le.inverse_transform(df['Iris.type'])
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(20,5))

plt.subplot(141)
sns.distplot(df[cols_names[0]], bins=20, color='blue')

plt.subplot(142)
sns.distplot(df[cols_names[1]], bins=20, color='blue')

plt.subplot(143)
sns.distplot(df[cols_names[2]], bins=20, color='blue')

plt.subplot(144)
sns.distplot(df[cols_names[3]], bins=20, color='blue')

plt.figure(figsize=(20,5))

plt.subplot(111)
plt.hist(df['Iris.type'], color='red')

plt.show()
```

```python
fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(141)
ax2 = fig.add_subplot(142)
ax3 = fig.add_subplot(143)
ax4 = fig.add_subplot(144)

sns.violinplot(x="Iris.cat_type", y=cols_names[0], data=df, split=True, ax=ax1)
sns.violinplot(x="Iris.cat_type", y=cols_names[1], data=df, split=True, ax=ax2)
sns.violinplot(x="Iris.cat_type", y=cols_names[2], data=df, split=True, ax=ax3)
sns.violinplot(x="Iris.cat_type", y=cols_names[3], data=df, split=True, ax=ax4)

plt.show()
```
